```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#probs = np.array([0.5, 0.5])
#probs = np.array([0.04, 0.04, 0.01, 0.01, 0.2, 0.2, 0.25, 0.25])
probs = np.array([1/8]*8)
probs
```

```python
# calculate self-information
self_info = -np.log2(probs)
self_info.shape
```

```python
self_info
```

```python
# plot the graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=probs, y=self_info, label='I(x) = -log2(P(x))')
plt.xlabel('Probability P(x)')
plt.ylabel('Self-information I(x)')
plt.title('Self-information vs Probability')
plt.grid(False)
plt.show()
```

```python
entropy = -(probs * np.log2(probs))
entropy
```

```python
sum(entropy)
```

```python
# plot the graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=probs, y=entropy, label='H(x) = -P(x) * log2(P(x))')
plt.xlabel('Probability P(x)')
plt.ylabel('Entropy H(x)')
plt.title('Entropy vs Probability')
plt.grid(False)
plt.show()
```
