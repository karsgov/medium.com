```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define probabilities
probs = np.linspace(0.0001, 1, 100)  # since log(0) gives runtime warning use a value closer to 0.
probs.shape
```

```python
# calculate self-information
self_info = -np.log2(probs)
self_info.shape
```

```python
# plot the graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=probs, y=self_info, label='I(x) = -log2(P(x))')
plt.xlabel('Probability P(x)')
plt.ylabel('Self-information I(x)')
plt.title('Self-information vs Probability')
plt.grid(False)
plt.show()
```
